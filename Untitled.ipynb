```python
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
conf = SparkConf().setMaster("local").setAppName("PySpark_feature_eng")
spark = SparkSession.builder.getOrCreate()
print(spark)
```

```python
from pyspark.sql import SQLContext
sqlContext = SQLContext(spark)
print(spark.version)
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
import os
print(os.listdir("/Users/kenxu/Desktop/Massive Data/Project"))
```

```python
train = pd.read_csv("/Users/kenxu/Desktop/Massive Data/Project/train.csv", nrows = 1000000)
test = pd.read_csv("/Users/kenxu/Desktop/Massive Data/Project/test.csv")
```

```python
train['key'] = pd.to_datetime(train['key'])
train['pickup_datetime']  = pd.to_datetime(train['pickup_datetime'])
train.head()
```

```python
for column in train.columns:
    train[column] = train[column].astype('str')
spark.conf.set("spark.sql.execution.arrow.enabled", "true")
df = sqlContext.createDataFrame(train)
```

```python
df.show()
```

```python
df.select(['fare_amount','pickup_datetime']).show()
```

```python
from pyspark.sql import functions as F
df1 = df.withColumn('key', F.split(df['key'], '\.')).withColumn('key', F.col('key')[0])
df2 = df1.withColumn('pickup_datetime', F.split(df['pickup_datetime'], '\+')).withColumn('pickup_datetime', F.col('pickup_datetime')[0])
df2.show()
```

```python
# Correct all the column types
df2 = df2.withColumn("key", to_timestamp("key"))
df2 = df2.withColumn("fare_amount", df["fare_amount"].cast("double"))
df2 = df2.withColumn("pickup_datetime", to_timestamp("pickup_datetime"))
df2 = df2.withColumn("pickup_longitude", df["pickup_longitude"].cast("double"))
df2 = df2.withColumn("pickup_latitude", df["pickup_latitude"].cast("double"))
df2 = df2.withColumn("dropoff_longitude", df["dropoff_longitude"].cast("double"))
df2 = df2.withColumn("dropoff_latitude", df["dropoff_latitude"].cast("double"))
df2 = df2.withColumn("passenger_count", df["passenger_count"].cast("bigint"))
df2.show()
```

```python
df3 = df2.filter(df["fare_amount"]>0)
df3[['fare_amount']].describe().show()
```

```python
# Plot distribution of pandas_df features
# Select a single column and sample and convert to pandas
sample_df = df3.select(['fare_amount']).sample(False, .0001, 12345)
pandas_df = sample_df.toPandas()
sns.distplot(pandas_df)
plt.xlabel('fare_amount') 
plt.show()
```
